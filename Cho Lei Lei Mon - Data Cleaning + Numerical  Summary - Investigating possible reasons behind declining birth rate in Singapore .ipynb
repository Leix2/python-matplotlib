## <font color=cornflowerblue>Data set 1: Median Age of grooms by ethnic group(annual) 

```python
import pandas as pd

#Read csv file
groomsAge = pd.read_csv('median-age-of-grooms-by-ethnic-group.csv', sep=',')

#Rename the column headers from random names to more meaningful names
ga = groomsAge.rename(columns={'year':'Year', 'level_1':'Ethnic Group','value':'Median Age of Grooms(Number of years)'})

#Display number of rows, number of columns and data type of each columns
print(ga.info())

print('--------------------------------------------------------------------------')

#Display numerical summary of all columns
print (ga.describe(include='all'))
print('--------------------------------------------------------------------------')

#Display numerical summary based on differt kinds in a group
#To see the variation of median age of grooms based on different type of ethnic groups
print( ga.groupby(["Ethnic Group"])[["Median Age of Grooms(Number of years)"]].describe())

print('--------------------------------------------------------------------------')

#Display the data set in a styled HTML representation
print("Styler : {}".format(type(ga.style)))
ga.style
```

## <font color=cornflowerblue>Data set 6: Distribution of monthly household expenditure by type of goods and services
